```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
import os
import glob
```

```python
filenames = glob.glob("dCG*.fits")
filenames.sort()
filenames = filenames[::-1]


fig = plt.figure(figsize=(10, 5), dpi=600)
for ii, filename in enumerate(filenames):
    spec = fits.open(filename)
    
    wavelength = spec[1].data['Lam']
    Lum = spec[1].data['Lum']
    index = np.argmin(np.abs(wavelength - 8000))
    NormLum = Lum / Lum[index]
    
    plt.plot(wavelength, NormLum+ii, c='k', lw=0.5)
    plt.text(8000, ii+1+0.1, filename.replace(".fits", ""), c='k')
```
